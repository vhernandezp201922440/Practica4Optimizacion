```python
#Función de reducción de Gauss para hallar las bases.
def reduccion_Gauss(B, b):
    n = len(b)
    x = np.zeros(n, float)
    for k in range(n-1):
        if np.fabs (B[k,k]) < 1.0e-12:
            for i in range(k+1, n):
                if np.fabs(B[i,k]) > np.fabs(B[k,k]):
                    B[[k,i]] = B[[i,k]]
                    b[[k,i]] = b[[i,k]]
                    break
        for i in range(k+1,n):
            if B[i,k] == 0:continue
            factor = B[k,k]/B[i,k]
            for j in range(k,n):
                B[i,j] = B[k,j] - B[i,j]*factor
            b[i] = b[k] - b[i]*factor
    x[n-1] = b[n-1] / B[n-1, n-1]
    for i in range(n-2, -1, -1):
        sum_ax = 0
        for j in range(i+1, n):
            sum_ax += B[i,j] * x[j]
        x[i] = (b[i] - sum_ax) / B[i,i]
    return B, x
```

```python
# Función de soluciones básicas con su base correspondiente.
def soluciones_basicas(A, b, solver='gauss'):
    vars = list(range(A.shape[1]))
    perm = list(itertools.combinations(vars, A.shape[0]))
    soluciones = {}
    soluciones['Solucion'] = []
    soluciones['Base'] = []
    soluciones['Variables básicas'] = []
    for case in perm:
        if np.linalg.det(A[:,case]) != 0:
            if solver == 'gauss':
                base, solucion = reduccion_Gauss(A[:, case], b) 
                soluciones['Base'].append(base)
            else:
                solucion = np.linalg.solve(A[:, case], b)
                soluciones['Base'].append(A[:, case])
            soluciones['Solucion'].append(solucion)
            soluciones['Variables básicas'].append(case)
        else:
            print("Matriz singular")
    return soluciones
```

```python
# Punto 2.b.1 , 2.b.2 y 2.b.3
# Función para encontrar todas las soluciones básicas usando la función del punto 1 
# ,verificar la factibilidad de las soluciones básicas encontradas
# y evaluar la funci+on de costo con cada una de las posibles 
# soluciones básicas factibles.
def soluciones_factibles(A, b):
    soluciones = soluciones_basicas(A,b, solver='linalg')
    soluciones['Tipo de solucion'] = []
    soluciones['Valor'] = []
    for i in range(len(soluciones['Solucion'])):
        solucion = soluciones['Solucion'][i]
        base = soluciones['Base'][i]
        case = soluciones['Variables básicas'][i]
        solucion_factible = True
        valor = 0
        for num in solucion:
            if num < 0:
                solucion_factible = False
        if solucion_factible:
            soluciones['Tipo de solucion'].append('Solución factible')
        else:
            soluciones['Tipo de solucion'].append('Solución no factible')
        solucion_completa = np.zeros((1,A.shape[1]))[0]
        for k in range(len(case)):
            solucion_completa[case[k]] = solucion[k]
        valor = -600*solucion_completa[0]-550*solucion_completa[1]-500*solucion_completa[2]
        soluciones['Valor'].append(valor)
    return soluciones
```

```python
# Punto 2.c.2
#Matriz A de las restricciones.
A = np.array([[20,18,15],[15,12,10],[200,220,250]])

#Matriz B de las igualdades de esas restricciones.
b = np.array([60,75,2000])

#Matriz C de las constantes de la función objetivo.
c = -np.array([600,550,500])

#Función linprog en donde se utiliza el método revised simplex y se muestra el reporte de optimización en consola.
print('Método: revised simplex')
start=time.time() # Tiempo inicial
res = linprog(c, A_ub=A, b_ub=b,method='revised simplex',options={"disp": True})
end=time.time() # Tiempo final

print('Optimal value:', round(res.fun, ndigits=2),
      '\nx values:', res.x,
      '\nNumber of iterations performed:', res.nit,
      '\nStatus:', res.message)

print("\nLas toneladas de carbón necesarias son: " + str(round(res.x[0])))

print("Las toneladas de petróleo necesarias son: " + str(round(res.x[1])))

print("Las toneladas de gas necesarias son: " + str(round(res.x[2])))

print("\nEl valor de la función objetivo es: " + str(round(-res.fun)))

print('\nTime spend:'+ str(end-start)) #Tiempo de ejecución de la función.


#Función linprog en donde se utiliza el método simplex y se muestra el reporte de optimización en consola.
print('\nMétodo: simplex')
start=time.time() # Tiempo inicial
res = linprog(c, A_ub=A, b_ub=b,method='simplex',options={"disp": True})
end=time.time() # Tiempo final

print('Optimal value:', round(res.fun, ndigits=2),
      '\nx values:', res.x,
      '\nNumber of iterations performed:', res.nit,
      '\nStatus:', res.message)

print("\nLas toneladas de carbón necesarias son: " + str(round(res.x[0])))

print("Las toneladas de petróleo necesarias son: " + str(round(res.x[1])))

print("Las toneladas de gas necesarias son: " + str(round(res.x[2])))

print("\nEl valor de la función objetivo es: " + str(round(-res.fun)))

print('\nTime spend:'+ str(end-start)) #Tiempo de ejecución de la función.

#Función linprog en donde se utiliza el método interior-point y se muestra el reporte de optimización en consola.
print('\nMétodo: interior-point')
start=time.time() # Tiempo inicial
res = linprog(c, A_ub=A, b_ub=b,method='interior-point',options={"disp": True})
end=time.time() # Tiempo final

print('Optimal value:', round(res.fun, ndigits=2),
      '\nx values:', res.x,
      '\nNumber of iterations performed:', res.nit,
      '\nStatus:', res.message)

print("\nLas toneladas de carbón necesarias son: " + str(res.x[0]))

print("Las toneladas de petróleo necesarias son: " + str(res.x[1]))

print("Las toneladas de gas necesarias son: " + str(res.x[2]))

print("\nEl valor de la función objetivo es: " + str(round(-res.fun)))

print('\nTime spend:'+ str(end-start)) #Tiempo de ejecución de la función.
```
